```python
import pandas as pd
import numpy as np
```

```python
# Load the Excel file
df = pd.read_excel("Cmeg_bacterial_data.xlsx", sheet_name="Trimmed")

# Save as CSV
df.to_csv("output.csv", index=False) 
```

```python
# Load the spreadsheet
df = pd.read_csv("output.csv")
```

```python
#replacing 'no's with 0, 'yes' with 1, and 'unknown' with Nan for computer manipulation
df["Pathogenic Species/Strains in Genus?"] = (df["Pathogenic Species/Strains in Genus?"]
    .replace({"no": 0, "No":0, "yes": 1, "Yes": 1, "sometimes": 2, "unknown":np.nan, "Unknown":np.nan,"unkown":np.nan,}))

df["Culturable?"] = (df["Culturable?"]
    .replace({"no": 0, "No":0, "yes": 1, "Yes": 1, "sometimes": 2}))
df["Plant Pathogen?"] = (df["Plant Pathogen?"]
    .replace({"no": 0, "No":0, "yes": 1, "Yes": 1, "sometimes": 2, "unknown":np.nan, "Unknown":np.nan,"unkown":np.nan,}))

df["Endosymbiont?"] = (df["Endosymbiont?"]
    .replace({"no": 0, "No":0, "yes": 1, "Yes": 1, "sometimes": 2, "unknown":np.nan, "Unknown":np.nan,"unkown":np.nan,}))

df["Genus Contains Antimicrobial Resistant Strains?"] = (df["Genus Contains Antimicrobial Resistant Strains?"]
    .replace({"no": 0, "No":0, "yes": 1, "Yes": 1, "sometimes": 2, "unknown":np.nan, "Unknown":np.nan,"unkown":np.nan,}))

df["Associated with Food-borne Illness?"] = (df["Associated with Food-borne Illness?"]
    .replace({"no": 0, "No":0, "yes": 1, "Yes": 1, "sometimes": 2, "unknown":np.nan, "Unknown":np.nan,"unkown":np.nan,}))

df["Disease Associations"] = (df["Disease Associations"]
    .fillna(0))


display(df.head(10))
```

```python
df.to_csv("output_fixed.csv", index=False) 
```
